```python
# importing necessary libraries
from time import process_time
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import random
import math
from scipy import sparse
from scipy.sparse import random
from scipy import stats
from numpy.random import default_rng
from sklearn.preprocessing import normalize
import numpy as np
import os
```

```python
# importing algorithm functions
import sys
sys.path.insert(1, '../functions')
from gk import gk
from mwrko import mwrko
from gko import gko
```

```python
# import data
import scipy.io
divorce = scipy.io.loadmat('../../data/divorce.mat')
photogram = scipy.io.loadmat('../../data/photogrammetry.mat')
stranke = scipy.io.loadmat('../../data/Stranke94.mat')
well = scipy.io.loadmat('../../data/well1033.mat')
```

### Divorce Dataset

```python
div = (divorce['Problem'][0][0]['A'])
div = div.toarray()
x = np.random.rand(9)
#div = normalize(div, axis=1, norm='l2')
y = div @ x
```

```python
gk_div = gk(div,x,y, 0.5*10**-5)
gko_div = gko(div,x,y, 0.5*10**-5)
mwrko_div = mwrko(div,x,y, 0.5*10**-5)
```

```python
plt.figure(figsize=(10,7))

plt.semilogy(np.arange(gk_div[0]), gk_div[1],color = 'blue', 
             linewidth = 5, linestyle = 'solid', label = 'GK')
plt.semilogy(np.arange(gko_div[0]), gko_div[1],color = 'red', 
             linewidth = 5, linestyle = 'dotted', label = 'GKO')
plt.semilogy(np.arange(mwrko_div[0]), mwrko_div[1],color = 'green',
             linewidth = 5, linestyle = 'dashed', label = 'MWRKO')

plt.xlabel('IT', fontsize = 15)
plt.ylabel('AR', fontsize = 15)
plt.title('Iterations vs. Approximation Error (Divorce Data)', 
         fontsize = 15, fontweight="bold")
plt.legend(fontsize="15")
```
